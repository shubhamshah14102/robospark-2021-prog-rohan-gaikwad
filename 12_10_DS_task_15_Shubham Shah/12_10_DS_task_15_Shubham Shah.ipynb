```python
import pandas as pd       # dataframe operations 
import numpy as np        # numeric operations
```

```python
path = '/content/drive/MyDrive/task15Data.csv'  # path of the dataset 

df = pd.read_csv(path)
df
```

```python
# drop irrelevant columns
df.drop(labels = {'passenger_id', 'ticket', 'fare', 'cabin' , 'boat', 'body', 'home.dest', 'name'} , axis = 1 , inplace = True)
df
```

```python
df.isnull().sum()
```

```python
col_null = ['age'] 
med = []

# loop for obtaining the median 
for i in col_null: 
  med.append(df[i].median())  

print("Median Values : " , med[0])

k = 0 

# loop for replacing null values 
for i in col_null:
  df[i].fillna(med[k] , inplace = True)    # fillna fills the null values 
  k = k + 1 

# fill embarked with max
df['embarked'].fillna('S', inplace = True)
```

```python
df.isnull().sum()
```

2. Normalize any required columns

```python
# Normalize age, parch and pclass
col_norm = ['age' ,  'pclass', 'parch']


max_ = [] 


# loop for getting the max of each column 
for i in col_norm : 
  max_.append(df[i].max())

k = 0 

# loop for dividing the column values with the max value 
for i in col_norm : 
  df[i] = df[i]/max_[k]
  k = k + 1 

df
```

3. Encode the categorical data [ use label encoding only for simplicity ]

```python
from sklearn.preprocessing import LabelEncoder

cols = ['sex', 'embarked']
df[cols] = df[cols].apply(LabelEncoder().fit_transform)

df
```

```python
df.corr()
```

```python
from sklearn import model_selection
```

4. Divide the data into train test split after separating the features from label

```python
X = df.drop('survived' , axis = 1) 
Y = df['survived']
```

```python
X_train , X_test , y_train, y_test = model_selection.train_test_split(X , Y ,test_size = 0.3 , shuffle = True , stratify=Y  )
```

5. Train a decision tree [ Sk Learn ] and calculate its score using score() [ find in decision tree documentation of Sk Learn ]

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
